```python
#import the pandas module
import pandas as pd
import os
```

```python
# Read all the csv files in the directory and concat the files one by one vertically

files=os.listdir()
data1=None
New=None
for i,f in enumerate(files):
    print(f)
    if f.split('.')[-1]=='csv':
        # Preprocessing before concatenation
        New=pd.read_csv(f)
        New=New.dropna()
        New=New.reset_index()
        New=New.drop(0,axis=0)
        if not isinstance(data1, pd.DataFrame):
            data1=New
            continue
        else:
            # Concatenate the data vertically
            data1=pd.concat([data1,New])
            
        
        
        
```

```python
data1
```

```python
data1=data1.drop('index',axis=1)
```

```python
data1
```

```python
data1.columns=['DateNtime','Contact','Destination','Duration','Type']
```

```python
# Save the combined data to a csv file as a checkpoint 
#checkpoint1
data1.to_csv('Combined.csv')
```

```python
# Clean the Destination string to get only the name of the destination
data1.Destination = data1.Destination.iloc[:].str.replace('to',' ').str.replace(' ','')
```

```python
# get all the date in dateTime format
data1['DateNtime']=pd.to_datetime(data1['DateNtime'])
```

```python
# Cleaning the Duration columns to get a numerical duration
data1.Duration=data1.Duration.str.replace('Min','')
```

```python
# Convert the cleaned Duration to 'int64'
data1.Duration=data1.Duration.astype('int64')
```
